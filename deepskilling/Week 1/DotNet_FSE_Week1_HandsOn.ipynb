```
using System;

// Step 2 & 3: Define a Singleton Class following the design pattern
public sealed class Logger
{
    // Create a class named Logger that has a private static instance of itself
    private static Logger _instance = null;
    
    // Thread safety lock object to ensure consistency across the application lifecycle
    private static readonly object _lockObj = new object();

    // Ensure the constructor of Logger is private to prevent external instantiation
    private Logger()
    {
        Console.WriteLine(">>>> Logger Instance Created Automatically <<<<");
    }

    // Provide a public static method to get the instance of the Logger class
    public static Logger GetInstance()
    {
        // Thread-safe implementation using double-check locking
        if (_instance == null)
        {
            lock (_lockObj)
            {
                if (_instance == null)
                {
                    _instance = new Logger(); 
                }
            }
        }
        return _instance;
    }

    // Utility method to simulate consistent logging functionality
    public void Log(string message)
    {
        Console.WriteLine($"[LOG - {DateTime.Now:yyyy-MM-dd HH:mm:ss}]: {message}");
    }
}

// Step 4: Create a test runner class to verify the Singleton implementation
public static class SingletonTestRunner
{
    public static void Run()
    {
        Console.WriteLine("--- Testing Singleton Pattern ---");

        // Requesting instances from different places in the code execution flow
        Logger logger1 = Logger.GetInstance();
        logger1.Log("First operation successful.");

        Logger logger2 = Logger.GetInstance();
        logger2.Log("Second operation successful.");

        // Verification step: Check if both point to the exact same instance in memory
        if (logger1 == logger2)
        {
            Console.WriteLine("\nSUCCESS: Both logger1 and logger2 point to the exact same instance!");
        }
        else
        {
            Console.WriteLine("\nFAILURE: Multiple instances exist.");
        }
    }
}

// Execute the test explicitly in your cell
SingletonTestRunner.Run();
```

```
using System;

// Step 2: Define Document Interface
public interface IDocument 
{
    string Open();
}

// Step 3: Create Concrete Document Classes matching each document type
public class WordDocument : IDocument
{
    public string Open() => "Opening Word Document... (.docx)";
}

public class PdfDocument : IDocument
{
    public string Open() => "Opening PDF Document... (.pdf)";
}

public class ExcelDocument : IDocument
{
    public string Open() => "Opening Excel Document... (.xlsx)";
}

// Step 4: Implement the Factory Method via an abstract base class
public abstract class DocumentFactory
{
    // The Factory Method
    public abstract IDocument CreateDocument(); 
}

// Concrete factory classes extending DocumentFactory to implement the method
public class WordFactory : DocumentFactory
{
    public override IDocument CreateDocument() => new WordDocument();
}

public class PdfFactory : DocumentFactory
{
    public override IDocument CreateDocument() => new PdfDocument();
}

public class ExcelFactory : DocumentFactory
{
    public override IDocument CreateDocument() => new ExcelDocument();
}

// Step 5: Create a test class to demonstrate object creation using the factory method
public static class FactoryTestRunner
{
    public static void Run()
    {
        Console.WriteLine("--- Testing Factory Method Pattern ---");

        // Create and open a PDF Document using the PDF Factory
        DocumentFactory pdfFactory = new PdfFactory();
        IDocument pdfDoc = pdfFactory.CreateDocument();
        Console.WriteLine(pdfDoc.Open());

        // Create and open a Word Document using the Word Factory
        DocumentFactory wordFactory = new WordFactory();
        IDocument wordDoc = wordFactory.CreateDocument();
        Console.WriteLine(wordDoc.Open());

        // Create and open an Excel Document using the Excel Factory
        DocumentFactory excelFactory = new ExcelFactory();
        IDocument excelDoc = excelFactory.CreateDocument();
        Console.WriteLine(excelDoc.Open());
    }
}

// Execute the test method explicitly
FactoryTestRunner.Run();
```

```
using System;
using System.Collections.Generic;

// Step 2: Setup the Product Class
public class Product
{
    public string ProductId { get; set; }
    public string ProductName { get; set; }
    public string Category { get; set; }

    public Product(string productId, string productName, string category)
    {
        ProductId = productId;
        ProductName = productName;
        Category = category;
    }
}

// Step 3: Implementation of Search Algorithms
public static class ECommerceSearchEngine
{
    // Linear Search Implementation - Time Complexity: O(N)
    public static Product LinearSearch(Product[] products, string targetId)
    {
        for (int i = 0; i < products.Length; i++)
        {
            if (products[i].ProductId == targetId)
            {
                return products[i]; // Found
            }
        }
        return null; // Not found
    }

    // Binary Search Implementation - Time Complexity: O(log N)
    // Note: This requires the array to be strictly sorted by ProductId beforehand
    public static Product BinarySearch(Product[] products, string targetId)
    {
        int left = 0;
        int right = products.Length - 1;

        while (left <= right)
        {
            int mid = left + (right - left) / 2;
            int comparison = string.Compare(products[mid].ProductId, targetId);

            if (comparison == 0)
            {
                return products[mid]; // Found
            }
            else if (comparison < 0)
            {
                left = mid + 1; // Discard left half
            }
            else
            {
                right = mid - 1; // Discard right half
            }
        }
        return null; // Not found
    }
}

// Test Class to execute and verify the operations
public static class SearchEngineTestRunner
{
    public static void Run()
    {
        Console.WriteLine("--- Testing E-commerce Search Function ---");

        // Set up unsorted array for linear search testing
        Product[] unsortedCatalog = new Product[]
        {
            new Product("P902", "Mechanical Keyboard", "Peripherals"),
            new Product("P102", "Wireless Mouse", "Peripherals"),
            new Product("P540", "Gaming Headset", "Audio"),
            new Product("P311", "HD Monitor", "Displays")
        };

        // Set up pre-sorted array for binary search testing (Sorted by ProductId ascending)
        Product[] sortedCatalog = new Product[]
        {
            new Product("P102", "Wireless Mouse", "Peripherals"),
            new Product("P311", "HD Monitor", "Displays"),
            new Product("P540", "Gaming Headset", "Audio"),
            new Product("P902", "Mechanical Keyboard", "Peripherals")
        };

        string targetId = "P540";

        // 1. Execute Linear Search
        Console.WriteLine($"\n[Action] Running Linear Search for Target ID: {targetId}...");
        Product resultLinear = ECommerceSearchEngine.LinearSearch(unsortedCatalog, targetId);
        if (resultLinear != null)
            Console.WriteLine($"[Success] Found via Linear: {resultLinear.ProductName} ({resultLinear.Category})");
        else
            Console.WriteLine("[Error] Product not found.");

        // 2. Execute Binary Search
        Console.WriteLine($"\n[Action] Running Binary Search for Target ID: {targetId}...");
        Product resultBinary = ECommerceSearchEngine.BinarySearch(sortedCatalog, targetId);
        if (resultBinary != null)
            Console.WriteLine($"[Success] Found via Binary: {resultBinary.ProductName} ({resultBinary.Category})");
        else
            Console.WriteLine("[Error] Product not found.");
    }
}

// Execute the code
SearchEngineTestRunner.Run();
```

```
using System;

public static class FinancialForecaster
{
    // Step 2 & 3: Implement a recursive algorithm to predict future value
    // Formula applied recursively: Future Value = Present Value * (1 + Growth Rate)
    public static double CalculateFutureValue(double presentValue, double growthRate, int periods)
    {
        // Base Case: If no years/periods remain, return the accumulated valuation
        if (periods <= 0)
        {
            return presentValue;
        }

        // Recursive Case: Compounding current value forward and decrementing periods left
        double compoundedValue = presentValue * (1 + growthRate);
        return CalculateFutureValue(compoundedValue, growthRate, periods - 1);
    }
}

// Test Runner Class to execute and verify the forecasting logic
public static class ForecastingTestRunner
{
    public static void Run()
    {
        Console.WriteLine("--- Testing Financial Forecasting Tool ---");

        // Input setup parameters
        double initialInvestment = 1000.00; // Starting capital (Present Value)
        double annualGrowthRate = 0.05;     // Constant yearly growth rate (5%)
        int trackingYears = 5;              // Predictive horizon (Periods)

        Console.WriteLine($"[Setup] Present Capital: ${initialInvestment}");
        Console.WriteLine($"[Setup] Projected Annual Growth: {annualGrowthRate * 100}%");
        Console.WriteLine($"[Setup] Forecasting Horizon: {trackingYears} Years\n");

        // Calculate recursive projection
        double forecastedValue = FinancialForecaster.CalculateFutureValue(initialInvestment, annualGrowthRate, trackingYears);

        // Round output for presentation readability
        Console.WriteLine($"[Result] Forecasted Asset Value after {trackingYears} years: ${Math.Round(forecastedValue, 2)}");
    }
}

// Execute the test explicitly
ForecastingTestRunner.Run();
```
